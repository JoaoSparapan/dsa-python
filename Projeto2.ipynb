## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
res = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()
res.head()
plt.figure(figsize = (15,5))
res.plot(x = 'Data_Pedido', y = 'Valor_Venda')
plt.show()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
res = df_dsa.groupby('Estado')['Valor_Venda'].sum().reset_index()
plt.figure(figsize = (15, 5))
plt.bar(res['Estado'], res['Valor_Venda'])
plt.xticks(rotation = 80)
plt.show()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
p1 = df_dsa.groupby('Cidade')['Valor_Venda'].sum()
top10 = p1.sort_values(ascending = False).head(10)
plt.bar(top10.index, top10)
plt.xticks(rotation = 80)
plt.show()
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df_dsa.head()
p1 = df_dsa.groupby('Segmento')['Valor_Venda'].sum()
plt.pie(p1, labels=p1.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Vendas por Segmento')
plt.show()
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.10)
df_dsa['Novo_Valor'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])
antes = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
depois = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Novo_Valor']
print('Média dos valores antes do desconto:', antes.mean(), ', após o desconto: ', depois.mean())
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
import locale
locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.strftime('%B')

p1 = df_dsa.groupby(['Ano','Mes','Segmento'])['Valor_Venda'].sum()
p1_unstacked = p1.unstack(level='Segmento')
p1_unstacked.plot(kind='line', marker='o')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.title('Vendas por Segmento ao longo do Tempo')

plt.legend(title='Segmento')
plt.xticks(rotation = 80)
plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
df_dsa_p10 = df_dsa.groupby(['Categoria','SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',ascending = False).head(12)
df_dsa_p10 = df_dsa_p10[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()

# Plotar o gráfico de barras horizontais
plt.barh(df_dsa_p10['Categoria'] + ' - ' + df_dsa_p10['SubCategoria'], df_dsa_p10['Valor_Venda'])

# Adicionar rótulos e título ao gráfico
plt.xlabel('Total de Vendas')
plt.ylabel('Categoria - SubCategoria')
plt.title('Top 12 Categorias e SubCategorias com Mais Vendas')

# Exibir o gráfico
plt.show()
```
